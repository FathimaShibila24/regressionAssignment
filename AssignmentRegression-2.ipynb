```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/car_age_price.csv')
```

```python
corr_metrix = data.corr()
plt.subplots(figsize = (10,4))
sns.heatmap(corr_metrix,linewidths=0.2,annot=True,cmap='coolwarm')
```

```python
x = data[['Year']]
y = data[['Price']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model = model.fit(x_train,y_train)
```

```python
pred =model.predict(x_test)
```

```python
pred
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is:\n',mean_squared_error(y_test,pred))
```

```python
year_2022 = np.array([[2022]])
predicted_price_2022 = model.predict(year_2022)[0]

print(f"Predicted Price for 2022 Model: â‚¹{predicted_price_2022[0]:.2f}")
```
